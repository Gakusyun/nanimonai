```python
# 
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties

%matplotlib inline
font = FontProperties(fname=r'\Users\gaoxjcn\AppData\Local\Microsoft\Windows\Fonts\LXGWWenKaiMono-Regular.ttf')

np.random.seed(1)
x1 = np.random.random(20)+1.5
y1 = np.random.random(20)+0.5
x2 = np.random.random(20)+3
y2 = np.random.random(20)+0.5

# 
plt.subplot(121)
plt.scatter(x1, y1, s=50, color='b', label='(+1)')
plt.scatter(x2, y2, s=50, color='r', label='(-1)')
plt.vlines(2.8, 0, 2, colors="r", linestyles="-", label='$wx+b=0$')
plt.title('', fontproperties=font, fontsize=20)
plt.xlabel('x')
plt.legend(prop=font)

# 
plt.subplot(122)
plt.scatter(x1, y1, s=50, color='b', label='(+1)')
plt.scatter(x2, y2, s=50, color='r', label='(-1)')
plt.scatter(3.5, 1, s=50, color='b')
plt.title('', fontproperties=font, fontsize=20)
plt.xlabel('x')
plt.legend(prop=font, loc='upper right')
plt.show()
```
